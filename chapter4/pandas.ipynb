# Hirarchical Indexing

So far, we have been dealing with Index objects that were a simple single value.

Hierarchical indexing uses MultiIndex objects, which are tuples of multiple values per Index.



```python
df = pd.DataFrame(np.random.randn(10,2),
                 index=[list('aaabbbccdd'),
                 [1,2,3,1,2,3,1,2,1,2]],
                 columns=['A','B'])

df
```

```python
# we can assign names to the MultiIndex object with the pandas.MultiIndex.names

df.index.names = ['alpha','numeric']; df
```

```python
# the pandas.DataFrame.reset_index() method removes all indexing levels
# from a MultiIndex DataFrame by default


df.reset_index()
```

```python
# the pandas.DataFrame.unstack() method has similar behavior and pivots the inner
# level of indexing and converts them to columns

df.unstack()
```

```python
# The pandas.DataFrame.stack() method does the opposite of unstack(...)


df.stack()
```

```python
# Let's examine teh structue of the MultiIndex DataFrame

df.stack().index
```

```python
# 1. Let's use the pandas.DataFrame.reset_index() to remove all hierarchical
# indexing from our previous df DataFrame

df = df.reset_index();df
```

```python
# Lets use the pandas.dataframe.groupby() method to group the A and B columns by the alpha column

grouped = df[['A','B']].groupby(df['alpha']); grouped
```

```python
# 3. we can use the dataframegroupby.describe() method to collect summary descriptive statistics

grouped.describe()
```

```python
# 4. we can apply the pandas.DataFrame.unstack() method using the 
# DataFrameGroupedBy.apply() method which accepts different functions
# and applies them to each group by the grouped object

grouped.apply(pd.DataFrame.unstack)
```

```python
# 5 There also exists the DataFrameGroupBy.agg() method
# which accepts functions and aggregates each column
# for each grorup by using that method.

grouped[['A','B']].agg('mean')
```

```python
# A similar method is the DataFrameGroupBY.transform() method
# with the only diffference being that transform works on one column
# at a time and returs a sequence of the same length as the series

from scipy import stats
grouped[['A','B']].transform(stats.zscore)
```

any missing data

# Discretization/bucketing of DataFrame values
The simplest way to achieve discretization is to create ranges of values and assign a single discreate
label to all values that fall within a certain bucket


```python
arr = np.random.randn(10)
arr
```

```python
 # The pandas.cut() method can be used to discretize these values,The following 
 # code uses the bins and labels=() arguments to bin the values into five discrete
 # values with the labels provided

cat = pd.cut(arr,bins=5,labels=['Very Low','Low','Med','High','Very High'])

cat
```

```python
# The pandas.qcut() method is similar but uses quartiles to bin the continous values to 
# to discrete values that each category has the same amount of observations

qcat = pd.qcut(arr,q=5,labels=['Very Low','Low','Med','High','Very High'])
qcat
```

```python
# The following cod block builds a pandas DataFrame consisting of the original continous
# values as well as the categories generated from cut and qcut

pd.DataFrame({'Value':arr,'Category':cat,'Quartile Category':qcat})
```

```python
# The pandas.Categorical.categories attribute provides us with the bucket ranges
pd.cut(arr,bins=5).categories
```

```python
# we can inspect the buckets for qcut as well
pd.qcut(arr,5).categories
```
